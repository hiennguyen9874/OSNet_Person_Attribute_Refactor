```python
import numpy  as np
import matplotlib.pyplot as plt

episode = 1e-10
```

```python
a = np.arange(episode, 1-episode, 0.01)
b = np.log(a)
plt.plot(a, b)
plt.show()
```

```python
a = np.arange(episode, 1-episode, 0.01)
b = -np.log(a)
plt.plot(a, b)
plt.show()
```

```python
a = np.arange(episode, 1-episode, 0.01)
b = np.exp(1-a)
plt.plot(a, b)
plt.show()
```

```python
a = np.arange(episode, 1-episode, 0.01)
b = np.exp(a)
plt.plot(a, b)
plt.show()
```
